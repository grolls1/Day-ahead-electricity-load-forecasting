**Forecasting class:**

```python
class Forecasting:

    def __init__(self):
        """Initialisation stage"""
        self.demand = None # (can be set to) electricity demand time series
        self.diff_steps = None # (can be set to) the differencing steps to perform
        self.lags = None # lags to use for the model specification
        self.endogenous = None
        # will be a dataframe containing the endogenous predictors (post-differencing, all lags)
                                                                            #if self.use_endogenous == 'True'
        self.exogenous = None # can be set to exogenous predictor frame previously formed
        self.diff_obs = 0
        self.lag_obs = 0
        self.use_endogenous = 'True'
        self.predictors = None # will be the full predictors set (endogenous and exogenous)
        self.target = None # electricity consumption - after differencing. 
        
        
    def preprocessing(self):
        """Set up X and y, using endogenous and exogenous variables."""
        

        self.endogenous = self.demand.copy()

        if self.diff_steps != None:

            self.diff_obs = sum(self.diff_steps)

            for d in self.diff_steps:
                self.endogenous = self.endogenous.diff(d).dropna() # think this is the same as doing it at the end


        if self.lags != None:

            self.lag_obs = max(self.lags)

            for lag in self.lags:

                self.endogenous[f'lag_{str(lag)}'] = self.endogenous['demand'].shift(lag)

            self.endogenous.dropna(inplace=True)

        else:
            self.use_endogenous = 'False'

            
            
        if str(type(self.exogenous)) != "<class 'NoneType'>":
            
            self.exogenous = self.exogenous.iloc[(self.diff_obs):]
            self.exogenous = self.exogenous.iloc[(self.lag_obs):]
            
            self.exogenous = pd.get_dummies(self.exogenous, columns=self.exogenous.columns, drop_first=True)

            
        if self.use_endogenous == 'True':
            self.predictors = pd.concat([self.endogenous, self.exogenous], axis=1)
            self.target = self.predictors.pop('demand')
        else:
            self.predictors = self.exogenous
            self.target = self.demand
        
    
            
    
    def predict(self, start, dynamic='day-ahead', model=LinearRegression(), end=None):

        if end == None:
            end = len(self.demand) # end = 139440   # this is meant to be the limit. I.e. if using loc, the end
                                                    # is really end-1.
            
            
        ##### FORECASTING (DIFFERENCED FORMAT) #####
        
        
            
        global predictions_diff
        predictions_diff = []
        ps = []
        predictions = pd.DataFrame([], columns=['demand'])

        for i in tqdm(range(start, end, 48)): # i is period 1 of each forecasted day

            model.fit(self.predictors.loc[:i], self.target.loc[:i])
            # using loc makes the start variable about the end forecasts. As in the undifferencing function.
            # Fit on X up to and including predictor set i (loc is inclusive).
            # Predictors set - the loc index corresponds to the differenced target variable being predicted.
            # Differenced series - the loc index corresponds to the latest demand obs within that entry.

            dynamic_predictors = self.predictors.loc[i:(i+47)].copy()
            # I think since loc is inclusive, this should actually be i+47. Wouldn't have hurt tho tbf.
            # The predictors set used spans the 48 periods, selected using loc.
            # With each forecast, the lag_1 entry for the next forecast will be updated with that forecast.
            # (what if there are other lags included which need incorporating? I.e. what if lag 336 is included?)
            # (then you'd also need to update the lag_48 entries... This needs fixing if i get there.)
            # (NO actually - since the predictions being made only go on 48 periods, it's only the lags up to)
            # (48 which will need updating. But it should be robust to lag_2 etc.)

            preds = []
            # preds store the forecasts made for each day-ahead section.
            for j in range(i, (i+48)):
            # j is the loc index of the observation being forecasted. Note: range is not inclusive - p 47 last one

                p = model.intercept_ + model.coef_.dot(dynamic_predictors.loc[j].values)
                # yep this is all good. Forming the forecast.

                
                if self.lags != None:
                    
                    lags_to_update = [lag for lag in self.lags if lag < 48]
                
                    if len(lags_to_update) != 0:
                        
                        for index in range(len(lags_to_update)):
                            
                            if j < (i + 48 - lags_to_update[index]):
                                
                                dynamic_predictors.iloc[:,index].loc[(j+1)] = p
                        
                        

                preds.append(p)

            predictions_diff.extend(preds)
            
            
            
            
            

            ##### UNDIFFERENCING #####
            ##### Note, this is done within each day-ahead iteration

            # want to put the 48 forecasts on the end of the less_differenced demand series.

            if self.diff_steps != None:
                diff_steps = self.diff_steps.copy()
                popped_diff_steps = []
                
                series = pd.DataFrame(preds, index=range(i, (i+48)), columns=['demand'])
                # series - initially the differenced forecasts from i to i+47.
                # (i+len(preds)) - but, len(preds) really should be 47, and if not, that's a problem needing fixing

                original_series = self.demand.copy()
                ##### This is surely only enough to form one fully differenced observation. Which actually i think is
                # right. The first less_differenced series, if diff_steps=[1,1,48], will be double-differenced.
                # So It will go from i-48 to i-1 (inclusive, so 48 obs). Which is what you want.

                # Original demand series. Using iloc since the range is not inclusive. i-1 is the last one.
                # The indexing here is crucial - is this the right series for the undifferencing?
                # Here, we need all the demand information which is contained within the forecasts.
                # For the first forecast, i: 
                        # if diff_steps = [1], then you need demand.iloc[(i-1):i]   - the previous demand obs.
                        # if diff_steps = [1,1], then you need demand.iloc[(i-2):i]   - the previous 2 obs.
                        # if diff_steps = [1,1,48], then you need demand.iloc[(i-50):i]   - the previous 50 obs.
                # This all seems set up fine.


                while len(diff_steps) > 0:
                # i.e. as long as the series has not been entirely undifferenced.

                    less_differenced = original_series.iloc[(i-sum(diff_steps)):i]

                    for d in diff_steps[:-1]:
                        less_differenced = less_differenced.diff(d).dropna()
                        # This series will have obs i-1 as its last observation, and i-sum(diff_steps) as the first

                        # less_differenced will still have the indices aligned:
                        # to undifference one layer (d) series.loc[100], you ONLY need less_differenced.loc[(100-d)]
                        # BUT - if obs (100-d) is not inside less_differenced, then you need to use series.
                        # NOTE - when using an obs from differenced series to undifference another differenced obs,
                        # is this fine? I think so. Since that first differenced obs contains all information known
                        # up to i, and the unknowns are filled in with the forecasts. So that's all the information 
                        # you can use to undifference the second obs. I do think this is fine.

                        # I think the if clause based on .loc[(j-d)] being inside less_differenced is the way to do this

                    d_step = diff_steps[-1]

                    if d_step == 1:

                        # getting the first entry in this series, from which we will perform the cumulative sum to go back one layer
                        first_entry = less_differenced.iloc[0].values[0]
                        # Note: if d_step==1, then there should only be one entry in this series.

                        # putting this entry into DataFrame format to allow concatenation with the (more) differenced series
                        first_entry_df = pd.DataFrame([first_entry], index=[(i-d_step-sum(popped_diff_steps))], columns=['demand'])
                        # (i-d_step-sum(popped_diff_steps)) : If diff_steps = [1,48]: we've done the undifferencing step for d=48.
                        # Which means that the j-48 term implicit within each j term has been removed, and placed at the beginning
                        # of the new series (which now starts at i-48). So, to do this d_step=1 undifferencing, we only need 
                        # demand observation i-49 (i-1-48) --- then we can do a simple cumsum.

                        # inserting this first entry as the first entry in the (more) differenced series
                        merged = pd.concat([first_entry_df, series])

                        # undo the layer of differencing with by taking the cumulative sum of the merged series
                        series = np.cumsum(merged).loc[i:]


                        popped_diff_steps.append(diff_steps.pop())

                        ### This d_step==1 stuff all seems fine. It will always use the first less_differenced term for 
                        ### a simple cumsum. 

                        ## Does the intuition get harder for d_step != 1?

                        ## I think for a robust function, the whole thing could be done in one.
                        # But it would need an if clause for each first_entry term. Easy with d_step=1 and
                        # with d_step greater than 48. But inbetween, there would have to be something more complex.
                        # Actually not an if clause, just that old range for the cumsum thing which i got rid of.
                        # The range would be used for differenced series.
                    else:

                        cum_sums = {}

                        for n in range(d_step):
                        # if d_step = 48, then range(0,48).

                            first_entry = less_differenced.iloc[n]['demand']
                            # first entry for the cumsum (with d_step=48, this is a simple addition)
                            first_entry_df = pd.DataFrame([first_entry], columns=['demand'], 
                                                          index=[(i+n-d_step-sum(popped_diff_steps))])
                            # (i+n-d_step-sum(popped_diff_steps)) : same intuition as before.

                            # need a try/except clause because of weirdness with lengths when dealing with diff(>1).
                            try:
                                differenced_series = series.iloc[[(n+j) for j in range(0,len(series),d_step)]]
                                # len(series) as (non-inclusive) end of j range in for loop: pretty intuitive. Need all of them
                            except:
                                differenced_series = series.iloc[[(n+j) for j in range(0,(len(series)-d_step),d_step)]]


                            merged_df = pd.concat([first_entry_df, differenced_series])

                            cum_sums[f'{str(i+n-d_step-sum(popped_diff_steps))}'] = np.cumsum(merged_df)

                        # same here - just a simple addition rather than cumsum.
                        series = pd.concat([cum_sums[str(i+n-d_step-sum(popped_diff_steps))] for n in range(d_step)]).sort_index().loc[i:]

                        popped_diff_steps.append(diff_steps.pop())

                ps.extend(series['demand'].values)

                predictions = pd.concat([predictions, series])
                
            else:
                
                ps.extend(preds)


        predictions_diff = pd.DataFrame(predictions_diff, 
                           index=list(range(start, start+len(predictions_diff))), columns=['demand'])

        return pd.DataFrame(ps, index=list(range(start, start+len(predictions_diff))),
                           columns=['demand'])
            
            

    
```

```python
index = []

mses_diff = []
maes_diff = []
medaes_diff = []

mses = []
maes = []
medaes = []

diff_steps = [None, [1], [1,1], [48], [1,48], [1,1,48], [1,1,48,336], [1,1,48,336,17520]]
lags = [[1], [1,2], [1,2,48], [1,48], [1,48,336], [1,2,48,336], [1,48,336,17520]]

start = (48*365*4)

for d in diff_steps:
    for l in lags:
        forecast = Forecasting()
        forecast.demand = demand.copy()
        forecast.diff_steps = d
        forecast.lags = l
        forecast.preprocessing()
        
        predictions = forecast.predict(start=start)
        p_diff = predictions_diff
        
        rmses.append(mean_squared_error(forecast.demand.loc[start:], predictions)**0.5)
        maes.append(mean_absolute_error(forecast.demand.loc[start:], predictions))
        medaes.append(median_absolute_error(forecast.demand.loc[start:], predictions))
        
        rmses_diff.append(mean_squared_error(forecast.target.loc[start:], p_diff)**0.5)
        maes_diff.append(mean_absolute_error(forecast.target.loc[start:], p_diff))
        medaes_diff.append(median_absolute_error(forecast.target.loc[start:], p_diff))
        
        index.append(f'd = {d}; l = {l}')
```

```python
results = pd.DataFrame([[a,b,c,d,e,f] for a,b,c,d,e,f in zip(rmses, maes, medaes, rmses_diff, maes_diff, medaes_diff)], 
            columns = ['rmse', 'mae', 'medae', 'rmse_diff', 'mae_diff', 'medae_diff'],
            index = index)
```

These coefficients are to be interpreted remembering that the series has been differenced, and that since the first dummy variable is dropped for each of the original variables (weekday, monthday, etc.), the coefficient for period 16 tells us the associative difference between differenced demand at period 16, and differenced demand at period 1. Thus the strong negative coefficient for period 16 suggests that the difference in electricity consumption between the half-hour commencing 7.30am and the half-hour commencing 7.00am tends to be lower than the difference in electricity consumption between the half-hour commencing 12.00am and the half-hour commencing 11.30pm. The fact that most periods have strong negative coefficients tells us simply that electricity consumption in the half-hour commencing 12.00am tends to be larger than consumption in the half-hour commencing 11.30pm, and that this difference is larger than differences in most other periods. Actually the opposite is true... Why negative?
